```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('../data_raw/2022-2023_bikethefts.csv', dtype={'LOR': str}, encoding='ISO-8859-1')
df.head()
```

```python
## Convert date columns to datetime, sort by date and reset index
df['TATZEIT_ANFANG_DATUM'] = pd.to_datetime(df['TATZEIT_ANFANG_DATUM'], dayfirst=True)
df['ANGELEGT_AM'] = pd.to_datetime(df['ANGELEGT_AM'], dayfirst=True)
df['TATZEIT_ENDE_DATUM'] = pd.to_datetime(df['TATZEIT_ENDE_DATUM'], dayfirst=True)
df.sort_values('TATZEIT_ANFANG_DATUM', inplace=True, ascending=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
# add TATZEIT_ANFANG_STUNDE to TATZEIT_ANFANG_DATUM as datetime
df['TATZEIT_ANFANG_DATUM'] = df['TATZEIT_ANFANG_DATUM'] + pd.to_timedelta(df['TATZEIT_ANFANG_STUNDE'], unit='h')
df['TATZEIT_ENDE_DATUM'] = df['TATZEIT_ENDE_DATUM'] + pd.to_timedelta(df['TATZEIT_ENDE_STUNDE'], unit='h')
```

```python
## drop irreleant columns
df.drop(['ANGELEGT_AM', 'VERSUCH', 'DELIKT', 'ERFASSUNGSGRUND', 'TATZEIT_ANFANG_STUNDE', 'TATZEIT_ENDE_STUNDE'], axis=1, inplace=True)
```

```python
## rename columns
df.columns = ['theft_start', 'theft_end', 'LOR', 'amount_of_damage_in_euro', 'type_of_bike']

## reorder columns
df = df[[
    'LOR',
    'theft_start',
    'theft_end',
    'type_of_bike',
    'amount_of_damage_in_euro'
]]

df
```
