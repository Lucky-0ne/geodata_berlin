```python
import pandas as pd
```

```python
df = pd.read_csv('index_covid.csv', sep=';')
df
```

```python
# te = df.dropna(how='any').copy()
df = df.drop(columns='id')
df['datum'] = pd.to_datetime(df['datum'])
df = df.sort_values(by='datum')
df['date'] = df['datum'].dt.year
df = df[df['date']==2024]
df = df.drop(columns=['datum', 'date'])
df = pd.DataFrame(df.sum()).reset_index()
df.columns = ['Bezirk', 'covid_count_2024_jan_feb']
df['Bezirk_norm'] = pd.Series(['Mitte', 'Friedrichshain-Kreuzberg', 'Pankow',
       'Charlottenburg-Wilmersdorf', 'Spandau', 'Steglitz-Zehlendorf',
       'Tempelhof-Schöneberg', 'Neukölln', 'Treptow-Köpenick',
       'Marzahn-Hellersdorf', 'Lichtenberg', 'Reinickendorf'])
df
```

```python
from geodata_berlin import LookupTableManager
```

```python
ltm = LookupTableManager('DISTRICTS', 'NAME')
result_df = ltm.map_geodata(df, 'Bezirk_norm')
result_df.head(3)
```

```python
result_df.explore('covid_count_2024_jan_feb')
```
