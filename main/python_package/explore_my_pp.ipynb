```python
import pandas as pd
import geodata_berlin as gdb
```

# PLZ:

```python
df = pd.DataFrame({'plzs': ['13353', '10115', '13355'], 'geile Spalte': ['Hood', 'test', 'Hier stinkts!']})
df
```

```python
ltm = gdb.LookupTableManager(resolution_mode='PLZ')
```

```python
result_df = ltm.get_meta_data(df, 'plzs')
result_df
```

```python
result_df.explore()
```
