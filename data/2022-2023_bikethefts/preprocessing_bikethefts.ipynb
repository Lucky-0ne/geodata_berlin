```python
import pandas as pd
```

```python
df = pd.read_csv('2022-2023_bikethefts.csv', dtype={'LOR': str}, encoding='ISO-8859-1')
df.head()
```

```python
## Convert date columns to datetime, sort by date and reset index
df['TATZEIT_ANFANG_DATUM'] = pd.to_datetime(df['TATZEIT_ANFANG_DATUM'], dayfirst=True)
df['ANGELEGT_AM'] = pd.to_datetime(df['ANGELEGT_AM'], dayfirst=True)
df['TATZEIT_ENDE_DATUM'] = pd.to_datetime(df['TATZEIT_ENDE_DATUM'], dayfirst=True)
df.sort_values('TATZEIT_ANFANG_DATUM', inplace=True, ascending=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
df['Year'] = df['TATZEIT_ANFANG_DATUM'].dt.year
df['Month'] = df['TATZEIT_ANFANG_DATUM'].dt.month
df['Day'] = df['TATZEIT_ANFANG_DATUM'].dt.day
df
```

```python
## save preprocessed data
# df.to_csv('2022-2023_bikethefts_preprocessed.csv', index=False)
```

```python
## get the number of bike thefts per month
df.groupby(['LOR', 'Year', 'Month']).size().reset_index(name='Bike_Thefts')
```

```python
## get the number of bike thefts per year
df.groupby(['LOR', 'Year']).size().reset_index(name='Bike_Thefts').sort_values('Year')
```
